## 层次聚类 (Hierarchical Clustering)

层次聚类是一种构建数据点层次结构的聚类分析方法。它不需要预先指定簇的数量，并且可以产生数据的多层次表示。

### 两种主要方法

1.  **凝聚式 (Agglomerative) 层次聚类 (自底向上):**
    *   开始时，每个数据点被视为一个单独的簇。
    *   在每个步骤中，算法会合并最相似的簇，直到只剩下一个簇或达到某个停止条件。

2.  **分裂式 (Divisive) 层次聚类 (自顶向下):**
    *   开始时，所有数据点都在一个簇中。
    *   在每个步骤中，算法将一个簇分裂成更小的簇，直到每个数据点形成一个簇或达到某个停止条件。

### 凝聚式层次聚类的步骤

1.  **计算相似度:**  计算所有数据点之间的相似度矩阵。常用的相似度度量包括欧氏距离、曼哈顿距离、余弦相似度等。
2.  **合并簇:**  找到最相似的两个簇并将它们合并。
3.  **更新相似度矩阵:**  重新计算新簇与其他簇之间的相似度。
4.  **重复步骤 2 和 3:**  直到所有数据点都在一个簇中或达到某个停止条件。

### 簇间距离的计算方法 (用于凝聚式聚类)

*   **单 linkage:**  簇间最近点之间的距离。
*   **全 linkage:**  簇间最远点之间的距离。
*   **平均 linkage:**  簇间所有点对之间距离的平均值。
*   **质心 linkage:**  簇的质心之间的距离。
*   **Ward linkage:**  合并簇后，簇内平方和增加的程度。Ward linkage 倾向于形成大小相似的簇。

### 层次聚类的优点

*   **无需预先指定簇的数量:**  可以观察不同层次的聚类结果，并根据实际情况选择合适的簇的数量。
*   **提供数据的层次结构:**  可以帮助理解数据点之间的关系。
*   **可视化:**  可以使用树状图 (dendrogram) 可视化聚类过程。

### 层次聚类的缺点

*   **计算复杂度高:**  特别是对于大型数据集。
*   **对噪声和异常值敏感:**  可能会影响聚类结果。
*   **难以撤销:**  一旦合并或分裂，就很难撤销之前的操作。

### 如何选择簇的数量

*   **观察树状图:**  在树状图中选择一个合适的切割点，将树分成若干个簇。
*   **使用评估指标:**  例如轮廓系数、Calinski-Harabasz 指数等。

### 总结

层次聚类是一种强大的聚类分析方法，可以用于发现数据的层次结构和关系。在选择层次聚类方法时，需要根据数据的特点和实际需求选择合适的相似度度量和簇间距离计算方法。

```python
from scipy.cluster import hierarchy
from matplotlib import pyplot as plt
%matplotlib inline
import pandas as pd

df = pd.read_csv("challenge-8-seeds.csv")

# 计算层次聚类
Z = hierarchy.linkage(df, method='ward', metric='euclidean')

# 绘制完整的树状图
plt.figure(figsize=(15, 8))
hierarchy.dendrogram(Z)
plt.show()

# 绘制截断的树状图
plt.figure(figsize=(15, 8))
hierarchy.dendrogram(Z, truncate_mode='lastp', p=15, show_leaf_counts=True, show_contracted=True)
plt.show()
```
