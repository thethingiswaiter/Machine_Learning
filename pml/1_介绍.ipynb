## 介绍

- 以概率视角看机器学习
- 将事件看作为随机变量, 并遵从某种特定分布

### 1 监督学习
- 学习一个从x到y的映射, 输出y是做好标签的
#### 1.1 分类
- 线性分类, 多分类, 输出是非线性的, 计算损失需要使用指示器函数
- 如果某一项较为重要, 也可以调整误差惩罚的权重, 比如某一分类的食物有毒, 需要谨慎判断
##### 1.1.1 经验风险最小化
- 其实就是损失函数的损失最小化, 可以调整损失的权重以及计算方式来优化学习
##### 1.1.2 条件概率分布
- 可以使用条件概率判断某一项分类输出的结果
##### 1.1.3 极大似然估计
- 在所有可能的参数取值中，选择使得观测到样本数据的概率最大的那个参数值作为估计值, 也就是p(y|x)
#### 1.2 回归
- 输出数据为实数, 拟合映射, 经常使用均值平方误差来判断
##### 1.2.1 线性回归
- 每个特征使用一维来调整参数, 有$f(x,\theta) = b + w_1x_1 + w_2x_2$
##### 1.2.2 多项式回归
- 特征之间的关系也会记录, 有$f(x,\theta) = b + w_1x_1 + w_2x_2 + w_3x_1x_2 + w_4x_1^2 + w_5x_2^2$
##### 1.2.3 深度神经网络
- 会在链路上递归的使用线性回归进行组合, 来表达更多样的特征关系
#### 1.3 过拟合和泛化
- 没有泛化能力的模型, 被称之为过拟合, 使用较复杂的模型匹配少量数据, 使得模型完美的映射了每一条训练数据
### 2 无监督学习
- 输入没有对应的输出来直接映射
- 针对大数据训练, 不需要耗费大代价去标记
#### 2.1 聚类
- 对内足够紧密, 对外足够远
#### 2.2 隐变量
- 经验性的假设隐变量为高斯分布
#### 2.3 自监督学习
- 使用辅助任务, 挖掘监督信息
#### 2.4 评估无监督学习
- 使用似然函数评估学习, 模型对数据的拟合程度越高, 说明解释能力越强
#### 2.5 增强学习
- 主打一个与环境交互的策略学习, 最大化积累奖励
### 3 数据
- 学习数据来源-见书
#### 3.1 处理离散数据
##### 3.1.1 one-hot编码
- 使用标签向量转化,如[1,0,0]
##### 3.1.2 特征交叉
- 将特征间两两或多个组合在一起相乘, 形成新的特征
#### 3.2 处理文本数据
- 需要注意超出字典的情况, 出现后可将其视为专有词汇处理
##### 3.2.1 词袋模型
- 忽略无意思词, 忽略词汇变形, 忽略词序, 只考虑词的频率, 模型简单易用, 但难以发现词义相关, 以及语序缺失导致的问题
##### 3.2.2 词频/逆文档频率
- 用于寻找相关性和分类, 越少见的词在文章中出现的频率越高, 越能代表该文章
##### 3.2.3 词嵌入
- 主动建立词汇之间相关量的矩阵, 补充简单模型间难以学习到词汇语义近似问题
#### 3.3 数据缺失
- 随机缺失, 可以直接干掉, 不会带来偏差, 可以以使用均值插补
- 非随机缺失, 需要建模进行预测,解释,比较难搞
